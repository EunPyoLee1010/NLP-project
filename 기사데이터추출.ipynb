```python
import pandas
import os
import re
from konlpy.tag import Okt
import pickle
```

```python
#okt 형태소 분석기
okt = Okt()
sentence = []
filename="ArticleData.csv"
name = ['date', 'category', 'press', 'title', 'content', 'url']
#pandas로 csv파일을 판다스 데이터프레임 형식으로 전환
df = pandas.read_csv(filename, sep=',', names = name, encoding='CP949')

#content에 해당되는 부분만 사용하고 해당 셀에 값이 없는 셀은 제거하여 새로운 변수로 저장
data_dropna = df['content'].dropna()
#중복 제거 후 리스트로 전환
dataset = list(set(data_dropna))

print(len(dataset))
```

```python
#기사 문장에서 명사, 동사, 형용사, 부사만 취하고 전부 제거하는 함수
def pos_article(dataset):
    index = 0
    for sent in dataset:
        if len(sent) == 0:
            continue
        sent = okt.pos(sent, stem = True)
        word = []
        for i in sent:
            if not i[1] == 'Noun' and not i[1] == 'Verb' and not i[1] == 'Adjective' and not i[1] == 'Adverb':
                continue
            if len(i[0]) == 1:
                continue
            word.append(i[0])
        word = ' '.join(word)
        word += '. '
        sentence.append(word)
        index += 1
        if index % 100 == 0:
            print(str(index) + " sentence is completed")
```

```python
pos_article(dataset)
print(sentence[0])
```

```python
#리스트 끝 인덱스의 공백과 단어의 마침표 제거
sent_list= []
for word in sentence:
    sent_list.append(word.split(' '))
print(sent_list[0])
for sent in sent_list:
    del sent[-1]
    sent[-1] = sent[-1].replace('.', '')
print(sent_list[0])
```

```python
# save
with open('data.pickle', 'wb') as f:
    pickle.dump(sent_list, f, pickle.HIGHEST_PROTOCOL)
```
