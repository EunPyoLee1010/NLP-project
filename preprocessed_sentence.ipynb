```
#모듈 임포트는 여기에서
import re
```

```
#html태그
cleantag = re.compile('<.*?>')
#특수문자
specialtext = '[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]'
#한글
hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')

#전처리 문장을 담을 리스트
pre_sent = []
with open("example.txt",'r') as f:
  lines = f.readlines()           #read로 받을수도 있음
  #전처리
  for line in lines:
    cleantext1 = re.sub(cleantag, '', line)
    cleantext2 = re.sub(specialtext,'', cleantext1)
    clean_sent = hangul.sub('',cleantext2)
    pre_sent.append(clean_sent)

#전처리된 문장 출력
for x in pre_sent:
  print(x)

#전처리된 문장을 다시 새로운 파일에 씀
#with open('preprocessed_sentence.txt', 'w') as f:
  
```
