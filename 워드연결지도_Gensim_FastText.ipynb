```python
import pickle
from gensim.models import FastText
from gensim.test.utils import common_texts
from gensim.test.utils import datapath
```

```python
#기사데이터에서 가져온 문장단어들 불러오기
with open('data.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
#https://fasttext.cc/docs/en/crawl-vectors.html
#위 사이트에서 한국어로 된 크롤링 pretrained_data 다운로드 후 불러오기
model = FastText.load_fasttext_format('Pretrained_data.bin', encoding='utf8')

#밑 메소드들은 다른방법으로 시도해봤었던 메소드들
#model = FastText.load("Pretrained_data.bin")

#맥에서 지원하는 모듈 
#cap_path = datapath("Pretrained_data.bin")
#model = FastText.load_facebook_model(cap_path)

#초기학습 모델
#model = FastText(size=300, window=5, min_count=1,sentences=data, iter=10)
```

```python
#불러온 데이터를 기존 학습 모델에 추가적으로 학습시키기
new_sentences = data

model.build_vocab(new_sentences, update=True)
model.train(new_sentences, total_examples=len(new_sentences), epochs=model.epochs)
```

```python
#New_model이라는 이름으로 모델 저장
model.save("New_model.bin")
```
